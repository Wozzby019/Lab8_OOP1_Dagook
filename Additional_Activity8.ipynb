```python
# Name: Jitse Jewel Thio-ac
# Student ID: TUPM-25-0827
# Activity 8: Object-Oriented Programming - Additional Exercises

# 1. Library Book Management System
class Book:
    def __init__(self, title, author, is_available=True):
        self.title = str(title)
        self.author = author
        self.is_available = is_available
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author):
        self.books.append(Book(title, author))

    def borrow_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                if book.is_available:
                    book.is_available = False
                    return f"Successfully borrowed '{book.title}'."
                return f"Error: '{book.title}' is already borrowed."
        return "Error: Book not found."

    def return_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                book.is_available = True
                return f"Successfully returned '{book.title}'."
        return "Error: Book not found."
    def display_status(self):
        print("\n--- Library Status ---")
        for book in self.books:
            status = "Available" if book.is_available else "Borrowed"
            print(f"Title: {book.title} | Author: {book.author} | Status: {status}")

# 2. Employee Payroll System
class Employee:
    def __init__(self, name, wage, hours):
        self.name = name
        self.wage = float(wage)
        self.hours = float(hours)
    def calculate_payroll(self):
        base_hours = min(self.hours, 160)
        ot_hours = max(0, self.hours - 160)
        base_salary = base_hours * self.wage
        ot_pay = ot_hours * (self.wage * 1.5)
        return base_salary, ot_pay

# 3. Online Shopping Cart System
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock
class ShoppingCart:
    def __init__(self):
        self.inventory = [
            Product("Laptop", 50000, 5),
            Product("Mouse", 500, 10),
            Product("Keyboard", 1200, 8)
        ]
        self.cart = []
    def add_to_cart(self, prod_name):
        for prod in self.inventory:
            if prod.name.lower() == prod_name.lower():
                if prod.stock > 0:
                    self.cart.append(prod)
                    prod.stock -= 1
                    return f"Added {prod.name} to cart."
                return f"Error: {prod.name} is out of stock."
        return "Error: Product not found."
    def view_cart(self):
        total = sum(item.price for item in self.cart)
        print("\n--- Shopping Cart ---")
        for item in self.cart:
            print(f"{item.name}: ₱{item.price}")
        print(f"Total Price: ₱{total}")

# --- TEST EXECUTION ---
def run_exercises():
    # Test Library
    print("--- Exercise 1: Library ---")
    lib = Library()
    lib.add_book("Python Basics", "Jewel")
    lib.add_book("Circuit Theory", "TUP Expert")
    lib.display_status()
    print(lib.borrow_book("Python Basics"))
    lib.display_status()
    # Test Payroll
    print("\n--- Exercise 2: Payroll ---")
    emp_name = "Jitse Jewel Thio-ac"
    wage, hours = 500, 170
    if wage < 0 or hours < 0:
        print("Invalid Input.")
    else:
        emp = Employee(emp_name, wage, hours)
        base, ot = emp.calculate_payroll()
        print(f"Employee: {emp.name}\nBase: ₱{base}\nOvertime: ₱{ot}\nTotal: ₱{base+ot}")
    # Test Shopping Cart
    print("\n--- Exercise 3: Shopping Cart ---")
    cart = ShoppingCart()
    print(cart.add_to_cart("Laptop"))
    print(cart.add_to_cart("Mouse"))
    cart.view_cart()
if __name__ == "__main__":
    run_exercises()
```
